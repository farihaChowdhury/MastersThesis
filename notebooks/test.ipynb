```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
from bokeh.plotting import show, output_file
```

```python
# Load the dataset
file_path = '../data/machine_temperature_system_failure.csv'
data = pd.read_csv(file_path)
```

```python
# Convert the 'timestamp' column to datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
# Drop the 'timestamp' column and prepare the features
X = data[['value']]

anomaly_periods = [
    ["2013-12-10 06:25:00.000000", "2013-12-12 05:35:00.000000"],
    ["2013-12-15 17:50:00.000000", "2013-12-17 17:00:00.000000"],
    ["2014-01-27 14:20:00.000000", "2014-01-29 13:30:00.000000"],
    ["2014-02-07 14:55:00.000000", "2014-02-09 14:05:00.000000"]
]

data.head()
```

```python
anomaly_data = pd.DataFrame()

# Filter data for each anomaly period and append it to the anomaly_data DataFrame
for start, end in anomaly_periods:
    start_time = pd.to_datetime(start)
    end_time = pd.to_datetime(end)
    anomaly_data = pd.concat([anomaly_data, data[(data['timestamp'] >= start_time) & (data['timestamp'] <= end_time)]])
```

```python
import matplotlib.pyplot as plt

# Calculate the number of anomaly points
anomaly_count = anomaly_data.shape[0]

# Calculate the number of normal points
normal_count = data.shape[0] - anomaly_count

# Data for visualization
labels = ['Normal Data', 'Anomaly Data']
counts = [normal_count, anomaly_count]

# Create a bar chart with updated colors (blue for normal, red for anomaly)
plt.figure(figsize=(8,6))
plt.bar(labels, counts, color=['blue', 'red'], alpha=0.6)
plt.title('Comparison of Normal and Anomaly Data Points')
plt.ylabel('Number of Data Points')

# Display the chart
plt.show()
```

```python
# Calculate the monthly minimum and maximum temperatures
data['month_year'] = data['timestamp'].dt.to_period('M')
monthly_min = data.groupby('month_year')['value'].min()
monthly_max = data.groupby('month_year')['value'].max()

bar_width = 0.35
index = np.arange(len(monthly_min))

plt.figure(figsize=(12, 6))

# Plot the monthly minimum temperatures as blue bars
plt.bar(index, monthly_min, bar_width, label='Min Temperature', color='blue', alpha=0.6)

# Plot the monthly maximum temperatures as red bars next to the min temperatures
plt.bar(index + bar_width, monthly_max, bar_width, label='Max Temperature', color='red', alpha=0.6)

# Add labels and title
plt.xlabel('Month-Year')
plt.ylabel('Temperature')
plt.title('Monthly Min and Max Temperature')

# Set the month labels for x-axis
plt.xticks(index + bar_width / 2, monthly_min.index.astype(str), rotation=45)

# Add a legend
plt.legend()

# Display the chart
plt.tight_layout()
plt.show()
```

```python
X_train, X_test = train_test_split(X, test_size=0.2, shuffle=False)
```

```python
# Apply Isolation Forest for anomaly detection (Unsupervised)
model = IsolationForest(n_estimators=300,contamination=0.01,   max_samples='auto')
model.fit(X_train)
```

```python
from sklearn.neighbors import LocalOutlierFactor


lof_model = LocalOutlierFactor(n_neighbors=500, contamination=0.07, novelty = True)
lof_ret = lof_model.fit(X_train)
```

```python
y_pred = lof_model.predict(X_test)
y_pred = np.where(y_pred == -1, 1, 0)
```

```python
from sklearn.svm import OneClassSVM


ocsvm_model = OneClassSVM(nu=0.2, gamma=0.005, kernel='rbf')
ocsvm_ret = ocsvm_model.fit(X_train)

y_pred = ocsvm_model.predict(X_test)
y_pred = np.where(y_pred == -1, 1, 0)
```

```python
import changefinder
import scipy.stats as stats

cf_model = changefinder.ChangeFinder(r=0.002, order=1, smooth=250)
ch_df = pd.DataFrame()
ch_df['value'] = X_train['value']
ch_df['anomaly_score'] = [cf_model.update(i) for i in ch_df['value']]
ch_score_q1 = stats.scoreatpercentile(ch_df['anomaly_score'], 25)
ch_score_q3 = stats.scoreatpercentile(ch_df['anomaly_score'], 75)
ch_df['anomaly_threshold'] = ch_score_q3 + (ch_score_q3 - ch_score_q1) * 3
ch_df['anomaly']  = ch_df.apply(lambda x : 1 if x['anomaly_score'] > x['anomaly_threshold'] else 0, axis=1)
```

```python

import matplotlib.pyplot as plt

# Assuming ch_df is already calculated and contains anomaly detection results
# Plot the values
plt.figure(figsize=(12, 6))

# Plot the normal data points
plt.plot(ch_df.index, ch_df['value'], label='Normal', color='blue', alpha=0.5)

# Highlight anomalies with red dots
anomalies = ch_df[ch_df['anomaly'] == 1]
plt.scatter(anomalies.index, anomalies['value'], color='red', label='Anomalies', marker='.')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Temperature Value')
plt.title('Detected Anomalies in Machine Temperature')

# Add legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
y_test = np.zeros_like(y_pred)
test_timestamps = data['timestamp'].iloc[X_test.index]

for start, end in anomaly_periods:
    mask = (test_timestamps >= pd.to_datetime(start)) & (test_timestamps <= pd.to_datetime(end))
    y_test[mask] = 1
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Output the results
print(precision)
print(recall)
print(f1)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.plot(test_timestamps, X_test['value'], label='Normal', color='blue', alpha=0.5)

anomalies = X_test[y_pred == 1]
anomalies_timestamps = data['timestamp'].iloc[anomalies.index]

print(anomalies)
plt.scatter(anomalies_timestamps, anomalies['value'], color='red', label='Anomalies', marker='.')


plt.xlabel('Timestamp')
plt.ylabel('Temperature Value')
plt.title('Detected Anomalies in Machine Temperature')
plt.legend()

plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix for LOF Anomaly Detection')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Show the plot
plt.tight_layout()
plt.show()
```
